```python
# train.csv, test.csv 데이터프레임의 결측치 처리하는 함수를 생성

def FillNa(df) :
    df['Age'] = train_df['Age'].fillna(df['Age'].mean())       # inplace = True : 원본데이터 반영
    df['Embarked'] = df['Embarked'].fillna('S')
    df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
    
    return df
```

```python
# 필요없는 열 삭제하는 함수

def dropColumn(df) :
    drop_li = ['Ticket', 'Name', 'Cabin', 'PassengerId']
    df.drop(drop_li, axis = 1, inplace = True)
    
    return df
```

```python
# object 형태의 컬럼을 인트형으로 바꾸는 함수
from sklearn.preprocessing import LabelEncoder

def labelEncoder(df) :
    obj = ['Sex', 'Embarked']
    for feature in obj :
        le = LabelEncoder()
        le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    
    return df
```

```python
# 최종 결과에 필요한 PassengerId를 삭제하기 전 따로 저장
passid = test_df['PassengerId']
```

```python
datas = [train_df, test_df]
for data in datas :
    FillNa(data)
    dropColumn(data)
    labelEncoder(data)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
# from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# 정확도 예측을 할 객체 변수 선언
label = train_df['Survived']
features = train_df.drop('Survived', axis = 1)


dt_clf = DecisionTreeClassifier()
lg_rg = LogisticRegression()

# X_train, X_test, y_train, y_test = train_test_split(features, label, test_size = 0.2, random_state = 11)

params = {
    'max_depth' : [2, 3, 5, 10],
    'min_samples_split' : [2, 3, 4],
    'min_samples_leaf' : [1, 5, 8]
}

grid_clf = GridSearchCV(dt_clf, param_grid = params, scoring = 'accuracy', cv = 5)
grid_clf.fit(features, label)

print(grid_clf.best_params_)                    # 최적 파라미터 추정
print(grid_clf.best_score_)                     # 최적 정확도
print(grid_clf.best_estimator_)                 # 최적 트리
```

```python
best_clf = grid_clf.best_estimator_

prd = best_clf.predict(test_df)
prd                                       # 이 array 를 csv 로 제출하면 답을 kaggle 에서 알려줌
```

```python
submission = pd.DataFrame()

submission['PassengerId'] = passid
submission['Survived'] = prd
submission
```
